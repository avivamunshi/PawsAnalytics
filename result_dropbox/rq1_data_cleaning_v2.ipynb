```python
df_tx = pd.read_csv('data/rq1/rq1TX_2years.csv')
df_wa = pd.read_csv('data/rq1/rq1WA_2years.csv')
df_tx_cat = pd.read_csv('data/rq1/rq1TX_cats_2years.csv')
df_wa_cat = pd.read_csv('data/rq1/rq1WA_cats_2years.csv')
print(df_tx.shape)
print(df_wa.shape)
print(df_tx_cat.shape)
print(df_wa_cat.shape)
```

```python
frames = [df_tx, df_wa, df_tx_cat, df_wa_cat]

df = pd.concat(frames)
print(df.shape)
print(df.head(10))
```

```python
df.drop_duplicates(subset=['id'], inplace=True)
print(df.shape)
```

```python
df_2 = df[['id', 'type', 'species', 'age', 'gender', 'size', 'coat', 'tags', 'name', 'description', 'status', 
         'published_at', 'status_changed_at',
         'colors.primary', 'attributes.spayed_neutered', 'attributes.house_trained', 'attributes.special_needs', 
         'attributes.shots_current', 'environment.dogs', 'contact.address.city',
       'contact.address.state', 'contact.address.postcode',
       'contact.address.country']]

print(df_2.columns)
print(df_2.shape)
```

```python
# Convert published_at and status_changed_at to datetime
df_2['published_at'] = pd.to_datetime(df_2['published_at'])
df_2['status_changed_at'] = pd.to_datetime(df_2['status_changed_at'])

# Get the month_year combination for the status_changed_at column
df_2['month_year'] = df_2['status_changed_at'].dt.strftime('%Y-%m')
```

```python
print(df_2.shape)
print(df_2.columns)
print(df_2)
```

```python
df2 = df_2[['id', 'type', 'contact.address.state', "month_year"]]
```

```python
print(df2.shape)
print(df2.columns)
print(df2)
```

```python
# Group by State, month_year, Type
grouped = df2.groupby(['contact.address.state', 'month_year', 'type']).size().reset_index(name='Count')

print(grouped.columns)
print(grouped.shape)
print(grouped)
```

```python
new_df = grouped.pivot_table(index=['contact.address.state', 'month_year'], columns='type', values='Count', aggfunc='sum').reset_index()
```

```python
new_df.columns
```

```python
new_df.columns.name = None
new_df = new_df.rename(columns={'Cat': 'Count of Adopted Cats', 'Dog': 'Count of Adopted Dogs'})
```

```python
print(new_df.shape)
print(new_df.columns)
print(new_df)
```

```python
new_df = new_df.dropna(subset=['Count of Adopted Dogs'])

print(new_df.shape)
print(new_df.columns)
print(new_df)
```

```python
new_df = new_df.dropna(subset=['Count of Adopted Cats'])

print(new_df.shape)
print(new_df.columns)
print(new_df)
```

```python
new_df.to_csv("data/final/adopted_pets_per_month.csv.", index=False)
```

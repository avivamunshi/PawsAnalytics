```python
import pandas as pd
import numpy as np
```

```python
dogs = pd.read_csv("Aviva/dogs_18_months.csv")
dogs['breeds.primary'] = dogs['breeds.primary'].str.lower()
dogs.head(5)
```

```python
dogs.columns
```

```python
dogs['breeds.primary'].unique()
```

```python
akc = pd.read_csv("akc_updated.csv")
akc.rename(columns = {'Unnamed: 0': 'Breed'}, inplace = True)
akc['Breed'] = akc['Breed'].str.lower()
akc.head(5)
```

```python
df_merged=pd.merge(dogs, akc, left_on='breeds.primary', right_on='Breed', how='left')
df_merged.head(5)
```

```python
df_merged.columns
```

```python
df_filtered = df_merged[['id', 'type', 'species', 'age', 'gender', 'size', 'coat', 'name', 'distance', 'breeds.primary', 'breeds.secondary', 'breeds.mixed','breeds.unknown', 'colors.primary', 'animal_type', 'distance','contact.address.city', 'contact.address.state', 'contact.address.postcode','contact.address.country','Breed', 'description_y','temperament', 'popularity', 'min_height', 'max_height', 'min_weight','max_weight', 'min_expectancy', 'max_expectancy', 'group','grooming_frequency_value', 'grooming_frequency_category','shedding_value', 'shedding_category', 'energy_level_value','energy_level_category', 'trainability_value', 'trainability_category','demeanor_value', 'demeanor_category']]
df_filtered.head(10)
```

```python
df_filtered.to_csv("dogs_18_joined_akc_pf_temp.csv", index=False, sep='|')
```
